```python
import numpy as np
import matplotlib.pyplot as plt
import os 
import cv2
import pandas as pd
import pickle
```

```python
DataDir = "/home/jjespper/masterIronhack/SEGUNDA_PRUEBA_CERVEZA/"

categories = ['Budweiser', 'Mahou', 'Pilsen_urquell','San_miguel']
```

```python
img_ancho = 300
img_alto = 300
```

```python
def create_training_data(categorylist,ancho,alto):
    lst=[]

    for category in categorylist:
        path = os.path.join(DataDir, category)
        class_num = categories.index(category)
        for img in os.listdir(path):
            try:
                img_array = cv2.imread(os.path.join(path,img))
                new_array = cv2.resize(img_array, (ancho, alto))
                new_array = cv2.cvtColor(new_array, cv2.COLOR_BGR2GRAY)
                plt.imshow(new_array)
                plt.show
                lst.append((new_array, class_num))
            except Exception as e:
                pass
    return lst
training_data = create_training_data(categories,img_ancho,img_alto)
```

```python
import random
random.shuffle(training_data)
```

```python
X=[]
y=[]
for features, label in training_data:
    X.append(features)
    y.append(label)
X = np.array(X).reshape(-1, 300, 300, 1)
```
